```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
final=pd.DataFrame()
for page in range(1,10):
    webpage = requests.get(f"http://quotes.toscrape.com/page/{page}/").text
    #print(f"http://quotes.toscrape.com/page/{page}/")
    soup=BeautifulSoup(webpage, 'lxml')
    
    quotes=[]
    authos=[]
    tags=[]
    quote_info=soup.find_all("div", class_="quote")
    for i in quote_info:
        quotes.append(i.find("span", class_="text").text.strip())
        authos.append(i.find("small", class_="author").text.strip())
        tg=[]
        for j in i.find_all("a", class_="tag"):
            tg.append(j.text.strip())
        tags.append(", ".join(tg))
    
    df = pd.DataFrame({"Quote": quotes, "Author":authos, "Tags": tags})
    final=final.append(df, ignore_index=True)
    
```

```python
final.to_csv("Quotes.csv", index=False)
```
